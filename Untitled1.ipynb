```python
import tensorflow as tf
import keras
import PIL.Image as img
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mnist = tf.keras.datasets.cifar10
(training_images,_), (val_images,_) = mnist.load_data()
training_images=training_images/ 255.0
val_images=val_images/ 255.0
```

```python
print(np.shape(training_images[1]))
print(type(training_images[1]))
plt.figure(figsize=(1,1))
plt.imshow(training_images[1])
```

```python
training_images=training_images.reshape(len(training_images),32,32,3)
val_images=val_images.reshape(len(val_images),32,32,3)
```

```python
model=keras.models.Sequential([
    keras.layers.Conv2D(32,(3,3),padding='same',activation='relu',input_shape=(32,32,3),name='first'),
    keras.layers.MaxPooling2D(2,padding='same'),

    keras.layers.Conv2D(16,(3,3),padding='same',activation='relu',name='second'),
    keras.layers.MaxPooling2D(2,padding='same'),
    
    #decoder network
    keras.layers.Conv2D(16,(3,3),activation='relu',padding='same',name='third'),
    keras.layers.UpSampling2D(2),
    
    keras.layers.Conv2D(32,(3,3),activation='relu',padding='same',name='fourth'),
    keras.layers.UpSampling2D(2),
    
    keras.layers.Conv2D(3,(3,3),activation='sigmoid',padding='same',name='last')
])
```

```python
model.summary()
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),loss='binary_crossentropy')
```

```python
history=model.fit(training_images,training_images,epochs=7,batch_size=256,validation_data=(val_images,val_images))
```

```python
pred=model.predict(val_images)
```

```python
plt.subplot(1,2,1)
plt.imshow(val_images[1])
plt.title("original")
plt.subplot(1,2,2)
plt.imshow(pred[1])
plt.title("reconstructed")
```

```python
plt.subplot(1,2,1)
plt.imshow(val_images[1])
plt.title("original")
plt.subplot(1,2,2)
plt.imshow(pred[1])
plt.title("reconstructed")
```

```python
original=(val_images[1]*255)
recons=(pred[1]*255)

original=np.squeeze(original).astype(np.uint8)
recons=np.squeeze(recons).astype(np.uint8)
```

```python
original=img.fromarray(original)
recons=img.fromarray(recons)
```
